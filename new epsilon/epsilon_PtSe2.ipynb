## nosym-cell, no soc

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('../src')
from PAOFLOW import PAOFLOW 
```

```python
ptse2 = PAOFLOW(savedir='./PtSe2/nosym/PtSe2-CdI2.save', model=None, outputdir='./PtSe2/nosym/output/', 
                          smearing='gauss', verbose=True)
```

```python
ptse2.dielectric_tensor(delta=0.1,intrasmear=0.05,
                          emax=8.,ne=801,d_tensor='all',degauss=0.001,from_wfc=True)
```

```python
epsi_xx = np.loadtxt('./PtSe2/nosym/output/epsi_xx.dat')
epsr_xx = np.loadtxt('./PtSe2/nosym/output/epsr_xx.dat')
epsqe_xx = np.loadtxt('./PtSe2/nosym/qe_output/epsxx.dat')
epsi_xy = np.loadtxt('./PtSe2/nosym/output/epsi_xy.dat')
epsr_xy = np.loadtxt('./PtSe2/nosym/output/epsr_xy.dat')
epsqe_xy = np.loadtxt('./PtSe2/nosym/qe_output/epsxy.dat')
epsi_zz = np.loadtxt('./PtSe2/nosym/output/epsi_zz.dat')
epsr_zz = np.loadtxt('./PtSe2/nosym/output/epsr_zz.dat')
epsqe_zz = np.loadtxt('./PtSe2/nosym/qe_output/epszz.dat')
epsi_yz = np.loadtxt('./PtSe2/nosym/output/epsi_yz.dat')
epsr_yz = np.loadtxt('./PtSe2/nosym/output/epsr_yz.dat')
epsqe_yz = np.loadtxt('./PtSe2/nosym/qe_output/epsyz.dat')

fig, ax = plt.subplots(2,2,figsize=(10,6))

ax[0,0].scatter(epsi_xx[:,0],epsi_xx[:,1],color='red',s=0.5,label="imag, PAOFLOW")
ax[0,0].scatter(epsr_xx[:,0],epsr_xx[:,1],color='black',s=0.5,label="real, PAOFLOW")
ax[0,0].plot(epsqe_xx[:,0],epsqe_xx[:,2],color='red',alpha=0.5,label="imag, QE",linewidth=1)
ax[0,0].plot(epsqe_xx[:,0],epsqe_xx[:,1],color='black',alpha=0.5,label="real, QE",linewidth=1)
ax[0,0].set_xlim([0,6])
ax[0,0].set_title("$\epsilon_{xx}$")
ax[0,0].set_ylim([-4,16])
ax[0,0].set_ylabel("Dielectric tensor")

ax[0,1].scatter(epsi_xy[:,0],epsi_xy[:,1],color='red',s=0.5,label="imag, PAOFLOW")
ax[0,1].scatter(epsr_xy[:,0],epsr_xy[:,1],color='black',s=0.5,label="real, PAOFLOW")
ax[0,1].plot(epsqe_xy[:,0],epsqe_xy[:,2],color='red',alpha=0.5,label="imag, QE",linewidth=1)
ax[0,1].plot(epsqe_xy[:,0],epsqe_xy[:,1],color='black',alpha=0.5,label="real, QE",linewidth=1)
ax[0,1].set_xlim([0,6])
ax[0,1].set_ylim([-1,1])
ax[0,1].set_title("$\epsilon_{xy}$")
ax[0,1].legend()

ax[1,0].scatter(epsi_zz[:,0],epsi_zz[:,1],color='red',s=0.5,label="imag, PAOFLOW")
ax[1,0].scatter(epsr_zz[:,0],epsr_zz[:,1],color='black',s=0.5,label="real, PAOFLOW")
ax[1,0].plot(epsqe_zz[:,0],epsqe_zz[:,2],color='red',alpha=0.5,label="imag, QE",linewidth=1)
ax[1,0].plot(epsqe_zz[:,0],epsqe_zz[:,1],color='black',alpha=0.5,label="real, QE",linewidth=1)
ax[1,0].set_xlim([0,6])
ax[1,0].set_ylim([-1,6])
ax[1,0].set_title("$\epsilon_{zz}$")
ax[1,0].set_xlabel("Energy (eV)")
ax[1,0].set_ylabel("Dielectric tensor")

ax[1,1].scatter(epsi_yz[:,0],epsi_yz[:,1],color='red',s=0.5,label="imag, PAOFLOW")
ax[1,1].scatter(epsr_yz[:,0],epsr_yz[:,1],color='black',s=0.5,label="real, PAOFLOW")
ax[1,1].plot(epsqe_yz[:,0],epsqe_yz[:,2],color='red',alpha=0.5,label="imag, QE",linewidth=1)
ax[1,1].plot(epsqe_yz[:,0],epsqe_yz[:,1],color='black',alpha=0.5,label="real, QE",linewidth=1)
ax[1,1].set_xlim([0,6])
ax[1,1].set_ylim([-1,1])
ax[1,1].set_title("$\epsilon_{yz}$")
ax[1,1].set_xlabel("Energy (eV)")
plt.show()
```
